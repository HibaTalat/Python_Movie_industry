```python
#We can see .0 in the end in both columns
df.loc[:,"gross"]

df.loc[:,"budget"]
```

```python
#change the data types to remove .0 from the end of both columns
#this will be done by first replacing NaN values with 0 
df['budget']=df['budget'].fillna(0)
df['budget']=df['budget'].astype('int64')
```

```python
df['gross']=df['gross'].fillna(0)
df['gross']=df['gross'].astype('int64')
```

```python
df.loc[:50,['released','year']]
#year in released column doesnot match year e.g.

df.loc[40,['released','year']]
```

```python
#Assuming released date has updated year 
#We change the dtype to string 
df['released']=df['released'].astype(str)
```

```python
#custom split the released into date and update year

df[['date','update_year']]=df['released'].str.split(',', expand=True)

#update year has country in it so using another delimiter to break the country and year into two
df[['update_year','country']]= df['update_year'].str.split('(', expand=True)
#cleaning the country by removing ) from the string using string replace
df['country'] = df['country'].str.replace(r')', '')

df
```

```python
df.sort_values(by=['gross'],inplace= False, ascending = False)
```

```python
#numerized the string values to show correlation of strinfs
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#another methode of conversion as i prefer label encoder
df_numerized1 = df

from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
for col_name in df_numerized1.columns:
    if(df_numerized1[col_name].dtype == 'object'):
        df_numerized1[col_name]= labelencoder.fit_transform(df_numerized1[:, 0])

df_numerized1
```

```python
df_numerized1.corr(method='pearson')
```

```python
#A complete correlation 
correlation_matrix = df_numerized1.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#correlation unstacking is a quick way to see comparison(pairing)

corr_mat = df_numerized1.corr()
corr_unstack= corr_mat.unstack()

corr_unstack
```

```python
#sorted pairs
corr_unstack=corr_unstack.sort_values()
corr_unstack
```

```python
high_corr = corr_unstack[(corr_unstack)>0.5]

high_corr.sort_values(ascending= True)
```
